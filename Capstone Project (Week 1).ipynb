# Final Capstone Project - Battle of the Neighbourhoods

## 1. Introduction
Many people under the age of 30 come to Canada for a short working holiday under an IEC Work Permit. When deciding which Canadian city to move to, a number of factors must be considered. Two of the most popular destinations for young work permit holders are Toronto and Vancouver. Both cities are attractive to young people as they offer an active nightlife, great restaurants and cafes and a number of popular tourist destinations. In addition, both cities are large enough to have a number of entry level jobs available at all times.

When first coming to Canada, many permit holders seek entry level jobs in restaurants and cafes. This is due to the accessibility of such positions, the fact that these positions offer tips as compensation and that in most cases, the worker may start their first shift almost immediately after being hired. This is especially important as both cities are quite expensive and it's easy to use a large portion of ones savings quickly without an income to supplement spending. It is, therefore, important for future travelers to know what kind of venues each city has and how many. In addition to that, it is important to know where the venues are located so that one may move into a neighbourhood which is expected to have a number of entry level jobs available. 

The aim of the following project is to analyze the types and number of venues associated with both Toronto and Vancouver in order to establish which city young work permit holders should move to.
